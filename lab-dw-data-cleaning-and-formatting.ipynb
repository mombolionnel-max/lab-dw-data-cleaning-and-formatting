```python
import pandas as pd

url="https://raw.githubusercontent.com/data-bootcamp-v4/data/main/file1.csv"

df=pd.read_csv(url)
```

```python
df
```

```python
 ## Exercise 1: Cleaning Column Names

columns=df.columns
df.columns=[column.strip().lower() for column in columns]

df.rename({"st":"state"}, axis=1, inplace=True)
```

```python
## Exercise 2: Cleaning invalid Values

def gender_change(value):
    if value == "Femal":
        return "F"
    elif value == "Male":
        return "M"        
    elif value == "female":
        return "F"
    elif value == "F":
        return "F"    
    elif value == "M":
        return "M"  
    else:
        return "nan"
    
df["gender"]=df["gender"].apply(gender_change)
```

```python
df["gender"].value_counts() 
```

```python
df
```

```python
my_state={"Cali"=="California", "WA"=="Washington", "AZ"=="Arizona" }

def translate_state(row):
    if row["state"] == "Cali":
        return "California"
    elif row["state"] =="WA":
        return "Washington"
    elif row["state"] =="AZ":
        return "Arizona"  
    else:
        return row["state"]
    
df["state"] = df.apply(translate_state, axis=1)
df
```

```python
df["state"].value_counts()
```

```python
def Bach(row):
    if row["education"] == "Bachelors":
        return "Bachelor" 
    else:
        return row["education"]
    
df["education"] = df.apply(Bach, axis=1)
df["education"].value_counts() 
```
